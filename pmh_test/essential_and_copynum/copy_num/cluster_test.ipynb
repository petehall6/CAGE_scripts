```python
import pandas as pd
import numpy as np

#Read in new essentiatllity csv and convert to hdf.  Need to set index to DepMap_ID to match current data set

new_df = pd.read_csv("CRISPRGeneEffect.csv")
new_df.rename(columns={new_df.columns[0]: "DepMap_ID"},inplace=True)
new_df.set_index('DepMap_ID',inplace=True)

print(new_df.shape)

```

```python
new_df.to_hdf("processed_ess_data_updated.h5", 'ess_df', mode='w')

new_df
```

```python
curr_df = pd.read_hdf("processed_ess_data_current.h5","ess_df")
curr_df
```

```python
combo_df = pd.concat([new_df,curr_df],axis=0,sort=False, join='outer')

combo_df.head()
print(combo_df.shape)
```

```python
combo_df.reset_index(inplace=True)

combo_df
```

```python
combo_df = combo_df.astype(str)

combo_df
```

```python
combo_df.drop_duplicates(inplace=True, subset=['DepMap_ID'])
print(combo_df.shape)

combo_df
```

```python
combo_df.columns = combo_df.columns.astype(str)
```

```python
cols = combo_df.columns

for col in cols:
    if 'A2M' in col:
        print(col)
```

```python
combo_df.set_index('DepMap_ID', inplace=True)
combo_df
```

```python
combo_df.to_hdf("cleaned_updated_processed_ess_data.h5", "ess_df", mode='w')

```
