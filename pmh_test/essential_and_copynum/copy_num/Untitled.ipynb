```python
import pandas as pd

raw_df = pd.read_csv("updated_copy_num.csv")
```

```python
raw_df
```

```python
cols = ['ModelID','StrippedCellLineName']
cells_df = pd.read_csv("Model.csv",usecols=cols)
cells_df.rename(columns={'ModelID':'DepMap_ID'},inplace=True)
cells_df
```

```python
merged_df = pd.merge(raw_df,cells_df,how='inner',on='DepMap_ID')
merged_df.insert(1,'cell_line',merged_df.pop('StrippedCellLineName'))
merged_df.drop(['Unnamed: 0','DepMap_ID'],axis=1,inplace=True)
```

```python
cell_sort = merged_df.sort_index(axis=0)
cell_sort.set_index('cell_line',inplace=True)
cell_sort.sort_index(axis=1,inplace=True)
cell_sort
```

```python
cell_sort.reset_index(inplace=True)
cell_sort
```

```python
cell_sort.to_hdf('copy_number_data_041825.h5','copy_num')
```
