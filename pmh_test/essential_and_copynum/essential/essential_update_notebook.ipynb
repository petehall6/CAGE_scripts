```python
import pandas as pd
```

```python
#Down load the following sets from DepMap.  https://depmap.org/portal/data_page/?tab=allData
#CRISPRGeneEffect.csv
#Model.csv
#CRISPRInferredCommonEssentials.csv
```

```python
#Create a model_df with the achilies index and cell_lines
```

```python
cols = ['ModelID','StrippedCellLineName']
model_df = pd.read_csv("Model.csv",usecols=cols)
model_df.rename(columns={'ModelID':'ach_index','StrippedCellLineName':'cell_line'},inplace=True)
model_df
```

```python
#create a essential_df of the common essential genes
```

```python
essential_df = pd.read_csv("CRISPRInferredCommonEssentials.csv")
essential_df.rename(columns={"Essentials":"essential_gene"},inplace=True)
essential_df['essential_gene'] = essential_df['essential_gene'].str.replace(r'\(.*\)','',regex=True).str.strip()
essential_df
```

```python
#create big df of all the essential scores
```

```python
chronos_df = pd.read_csv("CRISPRGeneEffect.csv")
```

```python
#Rename columns and remove the cell line index numbers in each column header ie (123)
```

```python
chronos_df.rename(columns={'Unnamed: 0':'ach_index'},inplace=True)
chronos_df
```

```python
#remove geneIDs and remove any whitespaces
```

```python
clean_cols = chronos_df.columns.str.replace(r'\(.*\)','',regex=True).str.strip()
print(clean_cols)

chronos_df2 = chronos_df.set_axis(clean_cols,axis=1)
print(chronos_df2)
```

```python
#Save each data frame to same .h5 file.  Key determines name of dataframe for I/O purposes.  Use 'a'ppend mode
```

```python
#fit cell_line to ACH
```

```python
chronos_cell = chronos_df2.merge(model_df, how='inner',on='ach_index')
chronos_cell.insert(1,"cell_line",chronos_cell.pop('cell_line'))
chronos_cell.drop(columns=['ach_index'],inplace=True)
chronos_cell
```

```python
#set cell_line as index for easier parsing down stream
```

```python
chronos_cell.set_index(['cell_line'],inplace=True)
chronos_cell
```

```python
ess_gene = essential_df
chronos_scores = chronos_cell
```

```python
#using append mode will allow adding multple data frames to single hdf.  
#***This will append to any current file or create a new one.  If making a new file be sure to delete any previously generated h5's. 
```

```python
ess_gene.to_hdf("processed_ess_data_041825.h5",key="ess_gene",mode='a')
```

```python
chronos_scores.to_hdf("processed_ess_data_041825.h5",key="chronos",mode='a')
```
