```python
new_df2 = new_df.set_index(['DepMap_ID'])

new_df2 = new_df2[sorted(new_df2.columns)]

new_df2
```

```python
usecols = ['ModelID','StrippedCellLineName']
cell_df = pd.read_csv('Model.csv',usecols=usecols)
cell_df.rename(columns={'ModelID': 'DepMap_ID'}, inplace=True)
cell_df

#cells = list(cell_df['StrippedCellLineName'])
#cells
```

```python
merged_df = new_df2.merge(cell_df, on='DepMap_ID', how='inner')
merged_df.head()
```

```python
merged_df.insert(1, 'StrippedCellLineName', merged_df.pop('StrippedCellLineName'))
merged_df = merged_df.rename(columns={'StrippedCellLineName': 'cell_line'})
merged_df = merged_df.drop(columns=['DepMap_ID'])
merged_df = merged_df.set_index('cell_line')
```
